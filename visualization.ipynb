```python
import torch
import argparse
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42	
```

```python
checkpoint = torch.load('checkpoints/Electricity_96_96_CycleNet_custom_ftM_sl96_pl96_cycle168_XLSTM_seed2024/checkpoint.pth', map_location=torch.device('cuda:0'))
weights = checkpoint['cycleQueue.data'].cpu().numpy()[:,119]
shifted_weights = np.roll(weights, 41)
plt.figure(figsize=(6, 4))
plt.plot(shifted_weights)
plt.axis('on') 
# Save the plot as a SVG file
plt.savefig('electricity[320].pdf', bbox_inches='tight', transparent=False)
plt.show()
```
